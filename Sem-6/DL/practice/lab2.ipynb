```python
import torch
import numpy as np

a = torch.tensor(5.0, requires_grad=True)
b = torch.tensor(7.0, requires_grad=True)

x = 2*a + 3*b
y = 5*a**2 + 3*b**3
z = 2*x + 3*y

print(z)
z.backward()
print(a.grad)

an_dz_da = 4 + 30 * a
print(an_dz_da)
```

```python
b = torch.tensor(7.0)
x = torch.tensor(2.0)
w = torch.tensor(8.0, requires_grad=True)

u = w*x
v = u+b
a = torch.relu(v)

print(a)
a.backward()
print(w.grad)

da_dw = x if v > 0 else 0
print(da_dw)
```

```python
b = torch.tensor(2.0)
x = torch.tensor(2.0)
w = torch.tensor(3.0, requires_grad=True)

u = w*x
v = u+b
a = torch.sigmoid(v)

print(a)
a.backward()
print(w.grad)

da_dw = a*(1-a)*x
print(da_dw)
```

```python
x = torch.tensor(1.2, requires_grad=True)

f = torch.exp(-x**2 - 2*x - torch.sin(x))

print(f)
f.backward()
print(x.grad)

da_dw = f*(-2*x - 2 - (torch.cos(x)))
print(da_dw)
```
